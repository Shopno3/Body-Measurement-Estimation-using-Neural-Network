```python
import cv2
import matplotlib.pyplot as plt
from pathlib import Path
from google.colab import files

# Upload the image manually
uploaded = files.upload()  # This will prompt you to upload an image

# Get the uploaded image filename
image_path = list(uploaded.keys())[0]

# Load the image
image = cv2.imread(image_path)

# Perform pose estimation on the image
results = model.predict(image, save=True)

# Display the annotated image
annotated_image_path = Path(results[0].save_dir) / "image0.jpg"
annotated_image = cv2.imread(str(annotated_image_path))

# Convert BGR to RGB for displaying with matplotlib
annotated_image_rgb = cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB)

# Plot the image
plt.imshow(annotated_image_rgb)
plt.axis('off')
plt.title(f"Annotated Image: {image_path}")
plt.show()

# Print the keypoints (pose information)
print("Keypoints Detected:", results[0].keypoints)
```

```python
# checking the names of the keypoints
keypoints = results[0].keypoints
if hasattr(keypoints, 'xy'):
    keypoint_names = [
        "nose", "left_eye", "right_eye", "left_ear", "right_ear",
        "left_shoulder", "right_shoulder", "left_elbow", "right_elbow",
        "left_wrist", "right_wrist", "left_hip", "right_hip",
        "left_knee", "right_knee", "left_ankle", "right_ankle"
    ]
    for idx, (coord, conf) in enumerate(zip(keypoints.xy[0], keypoints.conf[0])):
        print(f"{keypoint_names[idx]}: ({coord[0]}, {coord[1]}), Confidence: {conf.item()}")
```

```python
import math

# Function to calculate Euclidean distance
def calculate_distance(point1, point2):
    return math.sqrt((point2[0] - point1[0])**2 + (point2[1] - point1[1])**2)

# Extract keypoints coordinates for the first detected person
keypoints = results[0].keypoints

# Check if keypoints are available
if hasattr(keypoints, 'xy'):
    coordinates = keypoints.xy[0]  # Coordinates for the first person detected


# Distances in pixels
distances = {
        "shoulder_width": calculate_distance(coordinates[5], coordinates[6]),  # Between left_shoulder and right_shoulder
        "left_arm_length": (
            calculate_distance(coordinates[5], coordinates[7]) +  # Between left_shoulder and left_elbow
            calculate_distance(coordinates[7], coordinates[9])    # Between left_elbow and left_wrist
        ),
        "right_arm_length": (
            calculate_distance(coordinates[6], coordinates[8]) +  # Between right_shoulder and right_elbow
            calculate_distance(coordinates[8], coordinates[10])   # Between right_elbow and right_wrist
        ),
        "hip_width": calculate_distance(coordinates[11], coordinates[12]),  # Between left_hip and right_hip
        "torso_length": calculate_distance(coordinates[6], coordinates[12]),  # Between right_shoulder and right_hip
        "left_leg_length": (
            calculate_distance(coordinates[11], coordinates[13]) +  # Between left_hip and left_knee
            calculate_distance(coordinates[13], coordinates[15])    # Between left_knee and left_ankle
        ),
        "right_leg_length": (
            calculate_distance(coordinates[12], coordinates[14]) +  # Between right_hip and right_knee
            calculate_distance(coordinates[14], coordinates[16])    # Between right_knee and right_ankle
        ),
        "eye_distance(just for testing)": (
        calculate_distance(coordinates[1], coordinates[2]) #Between left and right eye
        ),
        "height": (
            calculate_distance(coordinates[1], coordinates[11]) + #total height for reference
            calculate_distance(coordinates[11], coordinates[15])
        )
}

# Print results
print(f"Shoulder Width: {distances['shoulder_width']:.2f} pixels")
print(f"Left Arm Length: {distances['left_arm_length']:.2f} pixels")
print(f"Right Arm Length: {distances['right_arm_length']:.2f} pixels")
print(f"Hip Width: {distances['hip_width']:.2f} pixels")
print(f"Torso Length: {distances['torso_length']:.2f} pixels")
print(f"Left Leg Length: {distances['left_leg_length']:.2f} pixels")
print(f"Right Leg Length: {distances['right_leg_length']:.2f} pixels")
print(f"Eye Distance (just for testing): {distances['eye_distance(just for testing)']:.2f} pixels")
print(f"Height(For Reference): {distances['height']:.2f} pixels")
```

```python
# Known real-world size of a reference object (height)
def get_reference_height_manual():
    """
    Get reference height from manual input
    """
    try:
        input_format = input("Enter height as: \n1. Total inches \n2. Feet and inches \nChoice (1 or 2): ")

        if input_format == "1":
            reference_height_inches = float(input("Enter height in inches: "))
        elif input_format == "2":
            feet = float(input("Enter feet portion of height: "))
            inches = float(input("Enter inches portion of height: "))
            reference_height_inches = (feet * 12) + inches
        else:
            print("Invalid choice. Using default height of 5'9\" (69 inches).")
            reference_height_inches = 69

    except ValueError:
        print("Invalid input. Using default height of 5'9\" (69 inches).")
        reference_height_inches = 69

    print(f"Using reference height: {reference_height_inches} inches")
    return reference_height_inches

reference_real_world_size = get_reference_height_manual()
```

```python
left_eye = coordinates[1]
left_hip = coordinates[11]
left_ankle = coordinates[15]
reference_pixel_distance = calculate_distance(left_eye, left_hip) + calculate_distance(left_hip, left_ankle)

# Calculate scaling factor
scaling_factor =  reference_real_world_size / reference_pixel_distance

# Convert distances to real-world measurements
real_world_distances = {
    measurement: value * scaling_factor for measurement, value in distances.items()
}

# Print real-world distances
for measurement, value in real_world_distances.items():
    print(f"{measurement}: {value:.2f} inches")

```

```python
# Extract only the required measurements
shoulder_width = real_world_distances["shoulder_width"]
left_arm_length = real_world_distances["left_arm_length"]
right_arm_length = real_world_distances["right_arm_length"]
torso_length = real_world_distances["torso_length"]
left_leg_length = real_world_distances["left_leg_length"]
right_leg_length = real_world_distances["right_leg_length"]

# Load the annotated image again
image_with_measurements = annotated_image.copy()

# Define text properties
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1
font_thickness = 2
color = (86, 252, 247)

# Define text positions near corresponding keypoints
positions = {
    "shoulder_width": ((coordinates[5] + coordinates[6]) / 2),  # Midpoint of shoulders
    "left_arm_length": ((coordinates[5] + coordinates[9]) / 2),  # Midpoint of left shoulder and wrist
    "right_arm_length": (coordinates[10]),  # right wrist
    "torso_length": ((coordinates[6] + coordinates[12]) / 2),  # Midpoint of right shoulder and right hip
    "left_leg_length": (coordinates[13]),
    "right_leg_length": ((coordinates[12] + coordinates[16]) / 2)
}

# Overlay text on the image
for measurement, position in positions.items():
    text = f"{measurement}: {real_world_distances[measurement]:.2f} in"
    position = (int(position[0]), int(position[1]))  # Convert to integer (pixel coordinates)
    cv2.putText(image_with_measurements, text, position, font, font_scale, color, font_thickness, cv2.LINE_AA)

# Convert BGR to RGB for Matplotlib display
image_with_measurements_rgb = cv2.cvtColor(image_with_measurements, cv2.COLOR_BGR2RGB)

# Display the final image with measurements
plt.figure(figsize=(25, 20))
plt.imshow(image_with_measurements_rgb)
plt.axis('off')
plt.title("Body Measurements Overlay")
plt.show()
```
